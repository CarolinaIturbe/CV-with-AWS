# Dividir lista en batches

```python
frentes = pd.read_csv("C:/Fotointrepretacion_Torreon/ML_results_ALL/results/result_all.csv", header = None)

frentes.head()
```

```python
frentes_predios = frentes[frentes[0] == 'frentes.predios']
frentes_predios
```

```python
keys_frentes_predios = frentes_predios[2]
list_keys_frentes = keys_frentes_predios.to_list()
```

```python
def divide_chunks(l, n):

    k, m = divmod(len(l), n)
    return (l[i*k+min(i, m):(i+1)*k+min(i+1, m)] for i in range(n))
 
 
batches = list(divide_chunks(list_keys_frentes, 20))

```

```python
for i in range(len(batches)):

    with open(f'C:/Fotointrepretacion_Torreon/ML_results_ALL/06_entregable/batches/batch_{i}.txt', 'w') as f:
      
        # using csv.writer method from CSV package
        write = csv.writer(f)
        
        #write.writerow(fields)
        write.writerows(batches[i])

```
